```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('mushroom.csv')
print(df.shape)
df.head()
```

```python
df['class'].value_counts()
```

```python
#from the above value counts we got to know that poisionous data have 1400 and edible column has 600 , which is imbalnce in data
```

```python
sns.countplot(data=df,x='class',orient='h',stat='count')
plt.show()
```

```python
df['class'] = df['class'].replace({'edible':1,'poisonous':0})
df.head()
```

```python
target = df['class']
target.head()
```

```python
np.unique(df['class'])
```

```python
features = df.drop(['class','Unnamed: 0'],axis=1)
features
```

```python
features.info()
```

```python
categorical_columns = features.select_dtypes(include=['object']).columns.tolist()
categorical_columns
```

```python
from sklearn import preprocessing
columns = categorical_columns # columns names where transform is required
for X in columns:
    exec(f'le_{X} = preprocessing.LabelEncoder()')  #create label encoder with name "le_X", where X is column name
    exec(f'features.{X} = le_{X}.fit_transform(df.{X})')  #execute fit transform for column X with respective lable encoder "le_X", where X is column name
features.head()  # to display transformed results
```

## import the model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,random_state=100,train_size=0.8)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(C=1.0,kernel='rbf',gamma=0.3)
```

```python
svc.fit(x_train,y_train)
```

```python
ypred = svc.predict(x_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
#we got 85% of accuracy from the randomly given samples C=1.0,kernal='rbf',gamma=0.3
```

```python
print(classification_report(y_test,ypred))
```

```python
#f1 score = eatable = 75% and poisionous = 90 as we have more no.of datas are poisiouns
```

```python
grid = GridSearchCV(svc,params,verbose=2)
```
