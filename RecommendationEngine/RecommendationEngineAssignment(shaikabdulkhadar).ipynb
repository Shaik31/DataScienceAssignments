```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('anime.csv')
print(df.shape)
df.head()
```

```python
df['name']
```

```python
def alphanum(element):
    return"".join(filter(str.isalnum, element))
```

```python
df['name'] = [alphanum(x) for x in df.name]
#df.loc[:,'alphanum'] = [alphanum(x) for x in df.col]
```

```python
df.isnull().sum()
```

```python
df['genre'] = df['genre'].fillna('Hentai')
```

```python
df['type'] = df['type'].fillna('TV')
```

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
```

```python
df.isnull().sum()
```

```python
df.groupby('genre')['rating'].mean().sort_values(ascending=False)
```

```python
df.groupby('genre')['rating'].count().sort_values(ascending=False)
```

```python
df.head()
```

```python
df1 = pd.pivot_table(data=df,index='anime_id',columns='genre', values='rating')
```

```python
df2 = df1.fillna(0,axis=1)
df2
```
