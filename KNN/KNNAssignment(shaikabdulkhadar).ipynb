## split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,train_size=0.8,random_state=10)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### import model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train,y_train)
```

```python
ypred = knn.predict(x_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
# we achieved 81% of the accuracy from the knnclassifer with the n_neighbours = 3
```

```python
print(classification_report(y_test,ypred))
```

```python
confusion_matrix(y_test,ypred)
```

```python
sns.heatmap(confusion_matrix(y_test,ypred),annot=True)
```

```python
#lets try with gridseach to get n_neighbours 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'n_neighbors':[1,2,3,4,5]}
```

```python
grid = GridSearchCV(knn,params,verbose=2)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
ypred = knn.predict(x_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
#after the gridsearch n_neighbors=1 we got 95% accuracy
```

```python
print(classification_report(y_test,ypred))
```

```python
confusion_matrix(y_test,ypred)
```
